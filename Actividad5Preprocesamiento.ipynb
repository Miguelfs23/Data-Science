5. ¿Tendrán los valores faltantes de `Weight` relación con alguna otra variable? Analízalo considerando las siguientes columnas:

Para ello deberás dividir cada columna en dos grupos: con/sin faltantes en `Weight` y utilizar métodos gráficos y la prueba t para ver si hay diferencias significativas. NO olvides incluir las conclusiones del análisis con los resultados obtenidos.

**NOTA**. Puedes reutilizar las funciones `Diagnose_MV_Numerical` y `Diagnose_MV_Categorical` definidas en Hands-On-Prep_Missing_Values.zip. Revisa nuevamente cómo utilizarlas.

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])

    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(10,4))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

6. Utiliza una medida de tendencia central para imputar los valores faltantes de `Weight`. ¿Cuál es recomendable y por qué?

```python
plt.figure(figsize=(8,5))
sns.histplot(response_df['Weight'], bins=20, kde=True, color='blue')

plt.axvline(response_df['Weight'].mean(), color='red', linestyle='dashed', linewidth=2, label="Media")
plt.axvline(response_df['Weight'].median(), color='green', linestyle='dashed', linewidth=2, label="Mediana")

plt.xlabel("Peso (kg)")
plt.ylabel("Frecuencia")
plt.title("Distribución del Peso con Media y Mediana")
plt.legend()
plt.show()

# Calcular la media y la mediana
mean_weight = response_df['Weight'].mean()
median_weight = response_df['Weight'].median()

# Seleccionar la mejor opción según la distribución
if abs(mean_weight - median_weight) < 5:  # Si la diferencia es pequeña, usar la media
    imputed_value = mean_weight
    method_used = "media"
else:  # Si la diferencia es grande, usar la mediana
    imputed_value = median_weight
    method_used = "mediana"

# Imputar valores faltantes
response_df['Weight'].fillna(imputed_value, inplace=True)

# Confirmar imputación
print(f"Valores faltantes imputados con la {method_used}: {imputed_value:.2f}")
```

7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=response_df['Weight'], color='skyblue')

plt.title("Boxplot de Weight")
plt.xlabel("Peso (kg)")
plt.show()

# Calcular cuartiles
Q1 = response_df['Weight'].quantile(0.25)
Q3 = response_df['Weight'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos
outliers = response_df[(response_df['Weight'] < lower_bound) | (response_df['Weight'] > upper_bound)]

# Contar valores atípicos
num_outliers = outliers.shape[0]

print(f"Número de valores atípicos en Weight: {num_outliers}")

outliers_df = outliers.copy()
outliers_df.to_csv('outliers_weight.csv', index=False)  # Guardar en CSV si es necesario

print("Valores atípicos almacenados en 'outliers_weight.csv'")

# Crear un nuevo DataFrame sin los valores atípicos
response_df_clean = response_df[(response_df['Weight'] >= lower_bound) & (response_df['Weight'] <= upper_bound)]

# Confirmar la eliminación
print(f"Nuevo tamaño del dataset después de eliminar outliers: {response_df_clean.shape[0]}")
```

9. ¿Hay encuestados que sean atípicos según su índice de masa corporal (`BMI`)? El `BMI` puede ser calculado usando la siguiente fórmula:

`BMI = Weight / Height ^ 2`

**NOTA**. El peso tiene que estar en kilogramos y la altura en metros, pero en el conjunto la altura se ha registrado en centímetros, por lo que deberás convertirla a metros.

```python
# Convertir la altura de cm a metros
response_df['Height_m'] = response_df['Height'] / 100

# Calcular el BMI
response_df['BMI'] = response_df['Weight'] / (response_df['Height_m'] ** 2)

# Mostrar los primeros valores
response_df[['Weight', 'Height', 'Height_m', 'BMI']].head()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(x=response_df['BMI'], color='purple')

plt.title("Boxplot del BMI")
plt.xlabel("Índice de Masa Corporal (BMI)")
plt.show()

# Calcular cuartiles
Q1 = response_df['BMI'].quantile(0.25)
Q3 = response_df['BMI'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos
outliers_bmi = response_df[(response_df['BMI'] < lower_bound) | (response_df['BMI'] > upper_bound)]

# Contar valores atípicos
num_outliers_bmi = outliers_bmi.shape[0]

print(f"Número de encuestados con BMI atípico: {num_outliers_bmi}")

outliers_bmi_df = outliers_bmi.copy()
outliers_bmi_df.to_csv('outliers_bmi.csv', index=False)  # Guardar en CSV si es necesario

print("Encuestados con BMI atípico almacenados en 'outliers_bmi.csv'")

# Crear un nuevo DataFrame sin los valores atípicos del BMI
response_df_clean = response_df[(response_df['BMI'] >= lower_bound) & (response_df['BMI'] <= upper_bound)]

# Confirmar la eliminación
print(f"Nuevo tamaño del dataset después de eliminar outliers de BMI: {response_df_clean.shape[0]}")
```
